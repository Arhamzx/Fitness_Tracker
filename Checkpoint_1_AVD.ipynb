#**BUSINESS UNDERSTANDING**

### BUSINESS OBJECTIVE

Dataset Fitness Tracker berisi data aktivitas fisik pengguna seperti jumlah langkah harian (*steps*), detak jantung (*heart rate*), jarak tempuh (*distance*), jenis aktivitas (*workout type*), serta jumlah kalori yang terbakar (*calories* *burned*).

Dalam dunia kesehatan, data ini sangat penting karena dibutuhkan oleh pengembang fitness tracker untuk memberikan rekomendasi aktivitas yang lebih relevan/ efektif pada penggunanya. Permasalahan utamanya adalah bagaimana mengukur dan mengevaluasi efektivitas berbagai jenis aktivitas fisik seperti langkah kaki, jenis workout maupun umur pada pembakaran kalori

#**ASSES SITUATION**

Dataset diperoleh dari platform Kaggle yang berisi data aktivitas fisik pengguna fitness tracker. Data ini mencakup variabel numerik seperti jumlah langkah, umur, maupun kalori yang terbakar, serta variabel kategori seperti jenis workout.

Beberapa tantangan yang mungkin terdapat pada analisis data ini adalah aktivitas fisik setiap individu cukup bervariasi, baik dari intensitas maupun duraasinya, sehingga dapat memengaruhi konsistensi pembakaran kalori. Selain itu, distribusi data pada jenis workout tertentu kemungkinan tidak seimbang dimana beberaoa jenis workout bisa cenderung lebih dominan dari yang lain, serta perlu diantisipasi adanya nilai yang hilang (*missing value*), data duplikat, maupun inkonsistensi pencatatan dari perangkat fitness tracker
