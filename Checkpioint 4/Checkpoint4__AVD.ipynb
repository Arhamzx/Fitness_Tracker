# **Mengecek Outliers**

```python
results = []

cols = Df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = Df[col].quantile(0.25)
  q3 = Df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = Df[(Df[col] < lower_bound) | (Df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(Df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_Df = pd.DataFrame(results)
results_Df.set_index('Kolom', inplace=True)
results_Df = results_Df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_Df)
```

## **Penanganan Outliers**

```python
columns_to_impute = ["Weight (kg)", "BMI"]

for col in columns_to_impute:
    Q1 = Df[col].quantile(0.25)
    Q3 = Df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    Df.loc[:, col] = Df[col].clip(lower=lower_bound, upper=upper_bound)
```

Outliers pada data ini ditangani dengan mengimputasi karena kurang dari 10%

## **Data Construction**

```python
def intensity_level(row):
    intensity_ratio = row['Avg_BPM'] / row['Max_BPM']

    if intensity_ratio < 0.60:
        return 'Low'
    elif 0.60 <= intensity_ratio <= 0.75:
        return 'Moderate'
    else:
        return 'High'

Df['Intensity_Level'] = Df.apply(intensity_level, axis=1)
```

Kolom Intensity_Level dibuat dengan mengelompokkan data berdasarkan rasio Avg_BPM dibagi dengan Max_BPM. Pengelompokan tersebut bertujuan untuk menyederhanakan data numerik menjadi kategori yang lebih mudah dianalisis, sehingga pola aktivitas fisik dapat diinterpretasikan dengan lebih jelas. Variabel ini juga lebih representatif dalam menggambarkan karakteristik latihan dibandingkan hanya menggunakan nilai detak jantung dalam bentuk angka mentah.

Berdasarkan boxplot diatas, terlihat bahwa median kalori terbakar pada setiap frekuensi latihan berada pada kisaran yang hampir sama, yaitu sekitar 1000–1050 kalori. Hal ini menunjukkan bahwa nilai tengah pembakaran kalori relatif konsisten di setiap tingkat frekuensi latihan dalam dataset. Selain itu rentang nilai kalori terbakar cukup luas, dengan nilai minimum sekitar 300 kalori dan maksimum mendekati 1800 kalori pada semua kategori frekuensi latihan. Ukuran interquartile range (IQR) pada setiap box juga terlihat hampir serupa, yang menunjukkan bahwa variasi pembakaran kalori antar frekuensi latihan tidak berbeda secara signifikan.

Berdasarkan boxplot yang menunjukkan bahwa median pembakaran kalori relatif sama pada setiap frekuensi latihan, fitness tracker dapat memberikan rekomendasi yang lebih berfokus pada kualitas latihan daripada hanya jumlah frekuensi latihan. Perangkat fitness tracker dapat mengarahkan pengguna untuk meningkatkan intensitas latihan, durasi aktivitas, atau memilih jenis workout tertentu agar pembakaran kalori menjadi lebih optimal.
