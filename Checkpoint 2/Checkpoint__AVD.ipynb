# **Mengecek Outliers**

```python
results = []

cols = Df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = Df[col].quantile(0.25)
  q3 = Df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

# **Comparison**

```python
import matplotlib.pyplot as plt
import seaborn as sns

kalori_workout = Df.groupby('Workout_Type')['Calories_Burned'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=kalori_workout.index, y=kalori_workout.values, palette='viridis')
plt.title('Rata-rata Kalori Terbakar Berdasarkan Jenis Latihan')
plt.xlabel('Jenis Latihan')
plt.ylabel('Rata-rata Kalori Terbakar')
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan hasil agregasi rata rata kalori terbakar berdasarkan jenis latihan, dapat disimpulkan bahwa setiap jenis workout memiliki tingkat efektivitas yang berbeda dalam membakar kalori. Workout dengan nilai rata rata kalori terbakar tertinggi seperti dapat dikategorikan sebagai latihan paling efektif, sedangkan yang lebih rendah seperti cenderung merupakan latihan dengan intensitas ringan hingga sedang.
